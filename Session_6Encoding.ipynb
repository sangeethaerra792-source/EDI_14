**LableEncoder**

- lableEncoder is a method will convert cat lables to num values
- It is under package: **sklearn(sickit-learn)**
     - The class name: **preprocessing**
         - The method name: **LablelEncoder**


```python
#fron<package>.<class>import<method>

from sklearn.preprocessing import LabelEncoder
```

- step-1: import the method
- step-2: save the method
- step-3: apply fit transform
- **fit tranform**
    - fit means find the logic
        - in above example fit means create **mapper** or **dict**
    - **transform** apply the logic and change the data
        - apply the **mapper on the data**

```python
file_path=r'C:\Users\SANGEETHA\OneDrive\Documents\Naresh IT\Data Files\Visadataset.csv'
visa_df=pd.read_csv(file_path)
visa_df
```
