**Step-13: Drop the column**

- labels: 'IndexLabel | None' = None :
  - 
- axis: 'Axis' = 0,
  -
- index: 'IndexLabel | None' = None,
  - 
- columns: 'IndexLabel | None' = None, 
- level: 'Level | None' = None,
- inplace: 'bool' = False,

```python
df.drop()
```

```python
df.drop(index=1)
```

```python
df.drop(columns='Marks')
```

```python
df.drop(index=1)
df.drop(columns='Marks')
# ========================================
df.drop('Marks')  # i fi dont give argument lable
df.drop(1)  # 1 become label
# ========================================
# we need to provide axis
# axis=0 indicates index
# axis=1 indicates columns
```

```python
df.drop('Marks',axis=0)
```

```python
df.drop('Marks',axis=1)
```

**inplace**

- when we do the operation the result output is not overwriting 
- means the original dataframe is same as it is.
- in order to do this we need use **inplace**
- by default inplace=False
- make **inplace = True** if you want overwrite the result

```python
df.drop('Marks',axis=1,inplace=True)
```
